```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
```

```python
df.head()
```

```python
df
```

```python
# Question 10
sns.boxplot(x = 'Category', y = 'Rating', data = df)
plt.xlabel('Category')
plt.xticks(rotation=90)
#plt.figure(figsize=(20,10))
plt.ylabel('Rating')
plt.title('Boxplot of App Ratings by Category')
plt.show()
```

```python
# Question 11
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#Question 12
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x="Size", y="Rating")
plt.xlabel('Size')
plt.xticks(rotation=90)
plt.yticks(rotation=90)

plt.ylabel('Rating')
plt.title('Sactterplot of Size by Rating')
plt.show()
```

```python
#Question 13
average_prices_by_type = df.groupby('Type')['Price'].mean().sort_values(ascending=True)

plt.figure(figsize=(10, 5))
average_prices_by_type.plot(kind='bar')

plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Price by App Type')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="Type", y="Price")
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.yticks(rotation=90)

plt.ylabel('Price')
plt.title('Barplot of Type by Price')
plt.show()
```

```python
# prompt: Create a bar chart to compare average prices by app type.

average_prices_by_type = df.groupby('Type')['Price'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=average_prices_by_type, x='Type', y='Price')
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.ylabel('Average Price')
plt.title('Average Price by App Type')
plt.show()
```
